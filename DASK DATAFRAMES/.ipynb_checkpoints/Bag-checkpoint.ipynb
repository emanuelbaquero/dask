```python
import dask.dataframe as df
data = df.read_csv('movie_rate_*.csv',sep='|',usecols=[i for i in range(1,34)])
```

```python
def reemplazar_llaves(co_str):
    inicio = co_str.find('{')
    final = co_str.find('}')
    return co_str[inicio+1:final]
```

```python
data.head(1)
```

```python
data['Descripcion'] = data.Description
```

```python
data['Descripcion'] = data.Descripcion.astype(str).apply(reemplazar_llaves)
```

```python
data.head(100)
```

```python
import dask.bag as db
```

```python
data_bag = data['Descripcion'].to_bag()
```

```python
data_bag_flatten = data_bag.map(lambda x: x.split(',')).flatten()
```

```python
data_bag_flatten.take(10)
```

```python
data_bag_flatten_clean = data_bag_flatten.map(lambda x: x.strip())
```

```python
data_bag_flatten_clean
```

```python
data_bag_frequencies = data_bag_flatten_clean.frequencies()
```

```python
data_bag_frequencies.take(100)
```
