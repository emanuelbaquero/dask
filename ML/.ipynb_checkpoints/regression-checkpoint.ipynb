```python
import dask.dataframe as df
import pandas as pd
import numpy as np
```

```python
from dask.distributed import Client, progress
client = Client(processes=False, threads_per_worker=4,n_workers=3,memory_limit='4GB')
client
```

```python
data = df.read_csv('movie_rate_*.csv',sep='|',usecols=[i for i in range(1,34)])
```

```python
daskDataCategorical = data.Name_x.astype('category').cat.as_known()
```

```python
daskDataDummies = df.get_dummies(daskDataCategorical) 
```

```python
ndata = df.concat([data,daskDataDummies],axis=1)
```

```python
ndataBudget = ndata[ndata.Budget.notnull()].persist()
```

```python
ndataBudget.head(10)
```

```python
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.model_selection import train_test_split
```

```python
X = np.array(ndataBudget.Budget)
```

```python
y = np.array(ndataBudget[['Adventure', 'Comedy','Western', 'Romance', 'Drama', 'Crime', 'Action', 'Mystery','Biography', 'Film-Noir', 'Horror', 'Musical', 'Sport', 'Thriller','Family', 'Fantasy', 'Sci-Fi', 'War', 'Animation', 'History', 'Music']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.externals import joblib

with joblib.parallel_backend('dask'):
    model.fit(X_train,y_train)
```
