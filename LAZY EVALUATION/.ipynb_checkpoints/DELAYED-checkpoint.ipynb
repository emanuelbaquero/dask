```python
from dask import delayed, compute
import numpy as np
```

```python
def do_something_1(x,y):
    return x+y + 2*x*y
def do_something_2(a,b):
    return  a**3 - b**3
def do_something_3(p,q):
    return p*p + q*q
```

# USING DASK.DELAYED

```python
final_result = []
```

```python
l_numeros_1 = [2,4,6,3,1,5,5]
l_numeros_2 = [4,6,5,7,8,4,2]
l_numeros_3 = [1,2,4,5,3,1,2]
```

```python
for i in range(0,len(l_numeros_1)):
    res_1 = delayed(do_something_1)(l_numeros_1[i],l_numeros_2[i])
    res_2 = delayed(do_something_2)(l_numeros_2[i],l_numeros_3[i])
    res_3 = delayed(do_something_3)(res_1,res_2)
    final_result.append(res_3)
```

```python
final_sum = delayed(sum)(final_result)
```

```python
final_sum.visualize()
```

```python
final_sum.compute()
```

# @delayed

```python
@delayed 
def do_something_1(x,y):
    return x+y + 2*x*y
@delayed 
def do_something_2(a,b):
    return  a**3 - b**3
@delayed 
def do_something_3(p,q):
    return p*p + q*q
@delayed
def suma(a):
    return np.sum(a)
```

```python
final_result = []
```

```python
fafa = [2,3,4]
```

```python
np.sum(fafa)
```

```python
for i in range(0, len(l_numeros_1)):
    res_1 = do_something_1(l_numeros_1[i],l_numeros_2[i])
    res_2 = do_something_2(l_numeros_2[i],l_numeros_3[i])
    res_3 = do_something_3(res_1,res_2)
    final_result.append(res_3)
```

```python
res_sum_lista = suma(final_result)
```

```python
res_sum_lista.compute()
```
