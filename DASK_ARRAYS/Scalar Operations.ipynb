```python
import dask.array as da
import pandas as pd
```

```python
my_arr = da.random.randint(10,size=10,chunks=3)
```

```python
my_arr.compute()
```

```python
mas_100 = my_arr+100
mas_100.compute()
```

```python
# REDUCTIONS
```

```python
my_arr.sum().visualize()
```

```python
dask_sum = my_arr.sum()
dask_sum.compute()
```

```python
my_ones_arr = da.ones((10,10),chunks=2,dtype=int)
```

```python
my_ones_arr.compute()
```

```python
my_ones_arr.mean(axis=0).compute()
```

```python
my_custom_array = da.random.randint(10,size=(4,4),chunks=(1,4))
```

```python
my_custom_array.visualize()
```

```python
my_custom_array.mean(axis=1).compute()
```

```python
# SLICING
```

```python
my_custom_array[1:3, 2:4].compute()
```

```python
my_small_arr = da.ones(4,chunks=2)
```

```python
my_small_arr.compute()
```

```python
brd_example1 = da.add(my_custom_array,my_small_arr)
```

```python
brd_example1.visualize()
```

```python
ten_array  = da.full_like(my_small_arr,10)
```

```python
ten_array.compute()
```

```python
brd_example2 = da.add(my_custom_array,ten_array)
```

```python
brd_example2.compute()
```

```python
#RESHAPING
```

```python
custom_arr_id = my_custom_array.reshape(16)
```

```python
custom_arr_id.compute()
```

```python
# STACKING
```

```python
stacked_arr = da.stack([brd_example1,brd_example2])
```

```python
stacked_arr.compute()
```

```python
stacked_arr_axis_1 = da.stack([brd_example1,brd_example2],axis=1)
```

```python
stacked_arr_axis_1
```

```python
# CONCATENATE
```

```python
concate_arr = da.concatenate([brd_example1,brd_example2])
```

```python
concate_arr.visualize()
```

```python
concate_arr = da.concatenate([brd_example1,brd_example2],axis=1)
```
