```python
x = da.random.random(10,chunks=2)
y = da.random.random(10,chunks=2)
```

```python
sum_x_y  = da.add(x,y)
```

```python
mean_x_y = da.mean(sum_x_y)
```

```python
mean_x_y.compute()
```

```python
mean_x_y.visualize()
```
